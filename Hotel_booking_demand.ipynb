```python
path = "/content/hotel_bookings.csv"
df = pd.read_csv(path)
```

```python
# 데이터 샘플 확인
df.head()
```

```python
# 각 칼럼의 속성 및 결측치 확인
df.info()
```

```python
# 칼럼의 통계치 확인
df.describe()
```

```python
# 특정 변수 분포 시각화
sns.distplot(df['lead_time'])
```
